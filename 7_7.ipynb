Napisz program symulujący spacer losowy i zliczający
moment pierwszego powrotu do punktu początkowego.

przeprowadź 1000 symulacji – jak wygląda ten rozkład?

czy program zawsze kończy swoje działanie, tzn. czy zawsze następuje powrót do punktu 0?

uwaga: komentarz na samym dole!

```python
import random
from collections import Counter
from operator import itemgetter


def getMove():
    return random.choice([1, -1])

N = 1000
wyniki = []
# 1000 symulacji
for n in range(N):
    position = 0
    position += getMove()
    counter = 1
    while position != 0:
        position += getMove()
        counter += 1
        if counter == 1000000:
            break
    wyniki.append(counter)
    
# print(*wyniki, sep='\n')
zliczoneDlugosci = Counter(wyniki)
```

Prawdopodobieństwa długości dróg:

```python
prawdopodobienstwa = {k: (v/1000) for k, v in zliczoneDlugosci.items()}
for k,v in sorted(prawdopodobienstwa.items(), key=itemgetter(1), reverse=True):
    print("długość ścieżki {} z prawdopodobieństwem {}".format(k, v))
```

komentarz: zdarzały się przypadki, że nie kończyło się liczyć po godzinie. Dlatego dodałem breakpoint (jeśli więcej niż 1000000 kroków, to stop). Ale gdyby program miał nieskończoną liość czasu na działanie, to zawsze by się skończył.
